### Ans.

Since $X$ has support $[e^{-1},\,1]$ and $Y = 1 + \ln(X)$, the range of $Y$ is $[0,\,1]$ because $\ln(X) \in [-1,0]$.

The map $x \mapsto 1 + \ln(x)$ is increasing, so for any $y$ we have

$$
F_Y(y) = \Pr(Y \leq y) = \Pr(1 + \ln X \leq y) = \Pr\!\big(X \leq e^{\,y-1}\big).
$$

Using the CDF of $X$, which is $F_X(x) = \ln(x) + 1$ on $[e^{-1},1]$, we plug in $x = e^{y-1}$:

$$
F_Y(y) = F_X(e^{y-1}) = \ln(e^{y-1}) + 1 = (y-1) + 1 = y,

valid for $0 \leq y \leq 1$.

So, in piecewise form:

- $F_Y(y) = 0$ for $y < 0$  
- $F_Y(y) = y$ for $0 \leq y \leq 1$  
- $F_Y(y) = 1$ for $y > 1$  

**Conclusion:** $Y$ follows a $\text{Uniform}(0,1)$ distribution.

**d.** Now let $U\sim U[0,1]$ be a uniformly distributed random variable on $[0,1]$, and consider the random variable $Z = e^{U-1}$. Compute the pdf $f_Z(z)$.

### Ans.

### Solution

Let $U \sim \text{Uniform}[0,1]$ and define $Z = e^{U-1}$. We want to find the pdf of $Z$.


**Step 1: Transformation formula**

Since $Z = g(U) = e^{U-1}$ is monotone increasing, we can use the standard formula for transformed variables:

$$
f_Z(z) = f_U(u) \left| \frac{du}{dz} \right|,
$$

where $u = g^{-1}(z)$.


**Step 2: Inverse transformation**

$$
z = e^{u-1} \implies \ln z = u - 1 \implies u = 1 + \ln z
$$


**Step 3: Derivative**

$$
\frac{du}{dz} = \frac{d}{dz}(1 + \ln z) = \frac{1}{z}
$$


**Step 4: Plug into pdf formula**

Since $U \sim \text{Uniform}[0,1]$, $f_U(u) = 1$ for $0 \le u \le 1$. So

$$
f_Z(z) = f_U(u) \left| \frac{du}{dz} \right| = 1 \cdot \frac{1}{z} = \frac{1}{z}.
$$


**Step 5: Support**

- When $U = 0$, $Z = e^{-1}$  
- When $U = 1$, $Z = 1$  

Hence, the support is $[e^{-1}, 1]$.


**Step 6: PDF of $Z$**

$$
f_Z(z) =
\begin{cases}
\dfrac{1}{z}, & e^{-1} \le z \le 1, \\
0, & \text{otherwise}.
\end{cases}
$$


**Step 7: Comparison with $X$ from part (a)**

The pdf $f_Z(z)$ is **exactly the same as $f_X(x)$** derived earlier.  
This confirms that $Z = e^{U-1}$ produces the same distribution as $X$, which makes sense because $Z$ is a monotone transformation of a Uniform[0,1] variable.

### Ans.

### Covariance of \(X\) and \(Y\)

The covariance is defined as:

$$
\text{Cov}(X,Y) = \mathbb{E}[XY] - \mathbb{E}[X]\mathbb{E}[Y].
$$

From the problem, we are given:

$$
\Pr(X=1 \text{ and } Y=1) = \frac{1}{3} \implies \mathbb{E}[XY] = \frac{1}{3}.
$$

Also, for Bernoulli variables:

$$
\mathbb{E}[X] = p, \quad \mathbb{E}[Y] = q.
$$

Plugging into the covariance formula:

$$
\text{Cov}(X,Y) = \mathbb{E}[XY] - \mathbb{E}[X]\mathbb{E}[Y] = \frac{1}{3} - p q.
$$

**Conclusion:**

$$
\text{Cov}(X,Y) = \frac{1}{3} - pq
$$

This is the covariance in terms of the success probabilities \(p\) and \(q\).

### Ans.

### Values of \(p\) and \(q\) for \(X\) and \(Y\) to be uncorrelated

$$
\text{Cov}(X,Y) = 0.
$$

From earlier, we have

$$
\text{Cov}(X,Y) = \frac{1}{3} - pq.
$$

Setting this to zero:

$$

**Conclusion:**

**c.** Are $X$ and $Y$ independent for the values in (b)?

---



### Ans.

### Independence of \(X\) and \(Y\) for the values in (b)

Two discrete random variables \(X\) and \(Y\) are independent if and only if

$$
\Pr(X=x, Y=y) = \Pr(X=x)\Pr(Y=y) \quad \text{for all } x, y.
$$

From the problem, we know

$$
\Pr(X=1, Y=1) = \frac{1}{3}.
$$

For the values in part (b), we have \(pq = 1/3\). We check the other joint probabilities:

\[
\begin{aligned}
\Pr(X=1, Y=0) &= p - \frac{1}{3}, \\
\Pr(X=0, Y=1) &= q - \frac{1}{3}, \\
\Pr(X=0, Y=0) &= 1 - p - q + \frac{1}{3}.
\end{aligned}
\]

Compare with the product of marginals:

\[
\begin{aligned}
\Pr(X=1)\Pr(Y=0) &= p (1-q) = p - pq = p - \frac{1}{3}, \\
\Pr(X=0)\Pr(Y=1) &= (1-p) q = q - pq = q - \frac{1}{3}, \\
\Pr(X=0)\Pr(Y=0) &= (1-p)(1-q) = 1 - p - q + pq = 1 - p - q + \frac{1}{3}.
\end{aligned}
\]

All joint probabilities match the product of marginals.

**Conclusion:**  

For the values in part (b) (\(pq = 1/3\)), \(X\) and \(Y\) are **independent**.

In Chapter $1$ example $1.4$, Grinstead and Snell talk about measuring two natural quantities associated with the so-called *simple random walk* $S_0 = 0, S_t = S_{t-1}+ X_t$, where $X_t = \pm 1$ with probability $\frac{1}{2}$ (they describe this as a simple betting game between Peter and Paul, where at each time $t$ they toss a fair coin, and if it comes up heads, then Peter wins a penny from Paul, else Paul wins a penny from Peter). We will now replicate their experiments.

**a.** Generate $n = 1000$ copies of a simple random walk $S^i = (S^i_0,S^i_1,\ldots,S^i_{T})$ (with $T=100$ and $i\in\{1,2,\ldots,n\}$). For each walk, compute $Z^i_T = \frac{S^i_T}{T}$ (i.e., the per-round winnings of Paul after $100$ games), as well as $W^i_T = \frac{1}{T}\sum_{t=1}^{T}\mathbb{1}_{\{S^i_t>0\}}$ (i.e., the fraction of rounds in $[1,T]$ where Paul is in the lead). Plot a histogram of both these quantities (using `plt.hist(data, density=True,bins='auto')`).


---

### Ans.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(42)
````

````python
def GenerateSimulationQ3(n, t):
  s = [[0 for i in np.arange(t)] for j in np.arange(n)]

  # Generate steps for all walks at once
  for i in np.arange(n):
    steps = np.random.rand(t)
    for j in np.arange(1, t):
      if steps[j] < 0.5:
        steps[j] = -1
      else:
        steps[j] = 1
      s[i][j] = s[i][j-1] + steps[j]

  # Compute Z_T
  z = [s[i][-1] / t for i in np.arange(n)]

  # Compute W_T
  # Check if S_t > 0 for t=1 to T (excluding S_0)
  w = [np.sum(1 if s[i][j]>0 else 0 for j in np.arange(t))/ t for i in np.arange(n)]

  plt.hist(z, density=True, bins='auto')
  plt.xlabel('Z_T (Per-round winnings)')
  plt.ylabel('Density')
  plt.title('Histogram of Z_T for n= ' + str(n) + ' T=' + str(t))
  plt.show()

  plt.hist(w, density=True, bins='auto')
  plt.xlabel('W_T (Fraction of time in lead)')
  plt.ylabel('Density')
  plt.title('Histogram of W_T for n= ' + str(n) + ' T=' + str(t))
  plt.show()
````

### Ans.

$\frac{\text{Volume of n-dim sphere}}{\text{Volume of n-dim cube with side 2}}  = P(\text{Point's distance from center} < 1)$


$\text{Volume of n-dim sphere} = 2^{dim} \times P(\text{Point's distance from center} < 1)$

````python
def GenerateNDimensionalPoints(n, dim):
  dimentions = [np.random.uniform(-1, 1, n) for i in np.arange(dim)]
  distance = np.sum([np.power(dimentions[i],2) for i in np.arange(dim)], axis=0)
  Z = distance<1
  return np.sum(Z)
````

We see that the estimate of the volume maximizes at ~5 and decreases after that. The estimate decreases significantly as we increase the dimension. This is expected as the volume of hypercube is significantly larger than the volume of the unit sphere when the dimensions increase. Since the unit circle occupies a very small volume of hypercube, it is difficult/arbitrary to estimate the probability of a given point lying in the unit circle, which is visible from the noisy estimate in the above plot.

This is also called the 'curse of dimensionality'. (used GPT to find the exact term)
